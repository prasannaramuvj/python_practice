```python
class House:
  def __init__(self,name):
    self.name = name

  def house_name(self):
    print(f"house name is {self.name}")

house1 = House("house1")
house2 = House("house2")

house1.house_name()


```

```python
class Employee:
  def __init__(self,name,amount):
    self.name = name
    self.amount = amount

  def show_balance(self):
    print(f"balance is {self.name} is {self.amount} ")

emp1 = Employee("prasanna",6777)

emp1.show_balance()
emp1.amount = 8888
emp1.show_balance()
```
