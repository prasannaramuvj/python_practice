```python
n = list(map(str,input().split()))

left = 0
right = len(n)-1

while left < right :
  n[left],n[right] = n[right],n[left]
  left += 1
  right -= 1

print(n)
```

```python
a =list(map(int,input().split()))
max1 = a[0]
max2 = a[0]

for i in range(0,len(a)):
  if a[i] > max1:

    max2 = max1
    max1 = a[i]
  elif a[i] != max1 and a[i] > max2:
    max2 = a[i]
print(max2)
```

```python
a = list(map(int,input().split()))
org =[]

for i in a:
  if i in org:
    continue
  else:
    org.append(i)
print(org)
```

```python
a = list(map(int,input().split()))
n = int(input())

rot = n % len(a)
p = []
for i in range(rot):
  l = a.pop(0)
  p.extend([l])

a.extend(p)
print(a)



```
