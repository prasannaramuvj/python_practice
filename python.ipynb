```python
arr = list(map(int,input().split()))
maxs = arr[0]
curr = arr[0]

for i in range(1,len(arr)):
  curr = max(arr[i],curr+arr[i])
  maxs = max(maxs,curr)
print(maxs)
```

```python
n = int(input())

transaction = []

for i in range(n):

  sender,receiver,amount,time = input().split()
  transaction.append([sender,receiver,int(amount),int(time)])

result = []

for i in range(n):
  s1,r1,a1,t1 = transaction[i]

  for j in range(i+1,n):
    s2,r2,a2,t2 = transaction[j]

    if s1==s2 and r1 == r2 and a1 == a2 and abs(t1-t2) <= 60:
       result.append(transaction[i])
       result.append(transaction[j])
dup = []

for i in result:
  if i not in dup:
      dup.append(i)
print(dup)
```
