```python
n = list(map(str,input().split()))

left = 0
right = len(n)-1

while left < right :
  n[left],n[right] = n[right],n[left]
  left += 1
  right -= 1

print(n)
```

```python
a = list(map(int,input().split()))
n = int(input())

rot = n % len(a)
p = []
for i in range(rot):
  l = a.pop(0)
  p.extend([l])

a.extend(p)
print(a)



```

```python
n = tuple(map(int,input().split()))
sum = 0

for i in range (0,len(n)):
  sum += n[i]
print(sum)
```

set unordered immutable not allow duplicate

```python
set1 = set(map(int,input().split()))
set2 = set(map(int,input().split()))
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)

```

```python
n = list(map(int,input().split()))
seen = set()
org = []
for i in n:
  if i not in seen:
    seen.add(i)
    org.append(i)
print(org)
```
